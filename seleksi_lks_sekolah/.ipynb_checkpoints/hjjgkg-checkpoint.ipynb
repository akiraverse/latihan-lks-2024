# Checking Outliers
From the analysis above, there is an extreme value in the sepal_length column, let's validate if all of the features have outliers or not?

#### we are going to check the outliers using z-score
z-score formula = data - mean / standard deviation

#### why z-score?
can be customized

```python
dfNumeric = df[['sepal_length','sepal_width','petal_length','petal_width']]
for i in dfNumeric.columns:
    print()
    print(i)
    print(np.std(dfNumeric[i]))
    std = np.std(dfNumeric[i])
    mean = np.mean(dfNumeric[i])
    zscore = dfNumeric[i] - mean / std
    print(zscore)
```

```python
# searching for outliers using z score
outliers = []

for column in dfNumeric.columns:
    quantile_75 = dfNumeric[column].quantile(q=0.75)
    
    # Identify outliers
    outliers.append(dfNumeric[dfNumeric[column] > quantile_75+1.5])
    
# Print outlier values and the index positions
outliers
```

# Making The Model, We Are Using KNN
Why using KNN?
- The problem is classification
- Small dataset: only 155 rows
- Low deminsional data, the dataset only has 4 features

```python
# Example of making predictions
from math import sqrt
 
def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((point1 - point2) ** 2))
# calculation the distance of X_test data point and X_train data point
```

```python
# calculating a X_test data point with all the X_train data points

def get_neighbors(X_train, y_train, test_instance, k):
    distances = []
    for i in range(len(X_train)):
        dist = euclidean_distance(test_instance, X_train[i])
        distances.append((dist, y_train[i]))
    
    # Sort the distances and get the nearest k neighbors
    distances.sort(key=lambda x: x[0])  # Sort by distance in ascending
    #  the distance is distance between the testing dataset and the training dataset
    # so each species has various distances (but obviously they are near one to another)
    # that's why it needs to be sorted so we can clasify which is the shortest distance of the data poin (X_test) to the species
    
    neighbors = [distances[i][1] for i in range(k)]  
    # Extract the labels of the k nearest neighbors
    # neighbors is a list of the train labels
    # it searches for the top 3 shortest distance with the data point (X_test)
    
    return neighbors
    # we are returning neighbors only because it will be used in the prediction
```

```python
def predict(X_train, y_train, X_test, k):
    predictions = []
    for test_instance in X_test:
        neighbors = get_neighbors(X_train, y_train, test_instance, k)
        # Majority vote: Find the most common class label among neighbors
        
        predicted_class = max(set(neighbors), key=neighbors.count) 
        # based on the neighbors, which is the species is the most frequent out of the three nearest neighbor
        # if there all of the three are different, then it will picked the first one
        
        predictions.append(predicted_class)
        # inserting the predicition to a 'predictions' list
    return predictions
```

```python
def accuracy(y_true, y_pred):
    correct = np.sum(y_true == y_pred)
    return correct / len(y_true) * 100
# calculating the accuracy in percentage
```

```python
# Set the number of neighbors
k = 3

# Make predictions
y_pred = predict(X_train, y_train, X_test, k)

# Calculate accuracy
acc = accuracy(y_test, y_pred)
print(f'Accuracy: {acc:.2f}%')


print("Predictions on test set:", y_pred_test, '\n')
print(y_test, '\n')
print(y_pred_test == y_test, '\n')
```

```python
# function for predicting new data point
def predict_single(X_train, y_train, new_instance, k):
    # Get the neighbors for this new instance
    neighbors = get_neighbors(X_train, y_train, new_instance, k)
    
    # Determine the most common class among neighbors (majority voting)
    predicted_class = max(set(neighbors), key=neighbors.count)
    return predicted_class
```
